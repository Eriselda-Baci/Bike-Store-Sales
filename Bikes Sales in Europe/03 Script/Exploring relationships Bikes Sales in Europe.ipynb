## Categorical Plots

    ## - Use a histogram to visualize the distribution of the variable 'Unit_Cost'.

```python
# Create categorical Plots for 'Unit_Cost'

sns.histplot(df['Unit_Cost'], bins = 10, kde = True)
```

```python
## Creating Categories: “Low cost” < 9 USD, “Medium cost” >= 9 USD and < 42 USD , “High cost” >= 42USD

df.loc[df['Unit_Cost'] < 9, 'Cost category'] = 'Low cost'
df.loc[(df['Unit_Cost'] >= 9) & (df['Unit_Cost'] < 42 ), 'Cost category'] = 'Medium cost'
df.loc[df['Unit_Cost'] >= 42, 'Cost category'] = 'High distance'
```

```python
## Check count for each categories

df['Cost category'].value_counts(dropna = False)
```

## - Use a histogram to visualize the distribution of the variable 'Unit_Price'

```python
# Create categorigal plots for the  'Unit_Price'

sns.histplot(df['Unit_Price'], bins = 10, kde = True)
```

```python
# Creating Categories: “Low price” < 150, “Medium price” >= 150 and <200 , “High price” >= 200

df.loc[df['Unit_Price'] < 10, 'Price category'] = 'Low price'
df.loc[(df['Unit_Price'] >= 10) & (df['Unit_Price'] < 370), 'Price category'] = 'Middle price'
df.loc[df['Unit_Price'] >= 70, 'Price category'] = 'High price'
```

```python
## Check count for each categories

df['Price category'].value_counts(dropna = False)
```
